## Modifying parameters

```julia
using MimiFUND
```

```julia
m_fund1 = MimiFUND.get_model()
m_fund2 = MimiFUND.get_model()
```

```julia
set_param!(m_fund2, :climatedynamics, :climatesensitivity, 4.5)
```

```julia
run(m_fund1)
run(m_fund2)
```

```julia
m_fund1[:climatedynamics, :climatesensitivity]
```

```julia
m_fund2[:climatedynamics, :climatesensitivity]
```

```julia
p1 = Mimi.plot(m_fund1, :climatedynamics, :temp)
p2 = Mimi.plot(m_fund2, :climatedynamics, :temp)

[p1 p2]
```

```julia
@vlplot(resolve={scale={y=:shared}}) + [p1 p2]
```
